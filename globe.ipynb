```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Create a globe projection
fig = plt.figure(figsize=(8, 8))
#ax = plt.axes(projection=ccrs.Orthographic(central_longitude=0.0, central_latitude=0.0))
ax = plt.axes(projection=ccrs.Orthographic(central_longitude=-5, central_latitude=10))

# Add coastlines and features
ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=1.5)
ax.add_feature(cfeature.STATES.with_scale('50m'))
ax.add_feature(cfeature.BORDERS, linestyle=":")
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='aliceblue')
ax.gridlines(draw_labels=False, linestyle='--', color='gray')

# Plot the red vertical line from lat 5 to 20 at lon = -5Â°
lons = [-5, -5]
lats = [5, 20]
ax.plot(lons, lats, color='red', linewidth=2, transform=ccrs.PlateCarree())

# Display the plot
plt.title('Globe with Lon/Lat Gridlines')
#plt.savefig("./globe.png")
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Create the figure and axis with Orthographic projection
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection=ccrs.Orthographic(central_longitude=0.0, central_latitude=0.0))

# Add built-in features (no downloading)
# Add coastlines and features
ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=1.5)
ax.add_feature(cfeature.STATES.with_scale('50m'))
ax.add_feature(cfeature.BORDERS, linestyle=":")
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='aliceblue')
ax.gridlines(draw_labels=False, linestyle='--', color='gray')

# Plot a vertical line at lon=355 (or -5) from lat=5 to 20
lons = [0, 360]
lats = [90, -90]
ax.plot(lons, lats, transform=ccrs.PlateCarree())

# Plot a vertical red line at lon=355 (or -5) from lat=5 to 20
lon = -5  # 355 degrees converted to -5
lats = [5, 20]
lons = [lon] * len(lats)
ax.plot(lons, lats, transform=ccrs.PlateCarree(), color='red', linewidth=2)

plt.title('Line on Globe: Lon=355Â°, Lat=5â20Â°', fontsize=12)
plt.show()
```
